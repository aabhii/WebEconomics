# CPM
Cost per thousand (CPM) is a marketing term used to denote the price of 1,000 advertisement impressions on one webpage. If a website publisher charges \$2.00 CPM that means an advertiser must pay \$2.00 for every 1000 impressions of its ad. 

```python
advertiser['CPM'] = (advertiser['cost']/advertiser['num_impressions'])
```

# eCPC
eCPC is effective cost per click in online advertising.

While CPC is the actual price for each click when buying clicks, eCPC is a calculated metric that tells you what the CPC would have been if you bought clicks instead of impressions, actions etc.

The term CPC is often used instead of eCPC in the everyday language.

# CPC
CPC is the Cost per click when buying CPC.

See the definition of a click.

When buying CPC from an ad network, the network usually converts the CPC bid to CPM (or RPM) in the background, via the historical or predicted Click-through rate, so the network can optimize their CPM. The ad networks inventory is after all ad impressions, which they want to get most out of.

In order to get a relatively high volume of ad impressions when buying CPC you need to make sure the ads is relevant and inviting to be clicked on. Low click-trough rates will make the ad network anticipate a low CPM for your campaign, which in turn gives the ad network a low revenue per impression, and they are likely to prioritize other competing campaigns.

```python
advertiser['eCPC'] = (advertiser['cost']/advertiser['num_clicks'])/1000
advertiser
```

```python
def get_variables_by_categorical(df, categorical):
    groupby_list = ['advertiser', categorical]
    advertiser = df.set_index(groupby_list)
    advertiser['num_impressions'] = df.groupby(groupby_list)['click'].count()
    advertiser['num_clicks'] = df.groupby(groupby_list)['click'].sum()
    advertiser['cost'] = df.groupby(groupby_list)['payprice'].sum()
    def click_cost(df):
        return df.loc[df['click'] == 1, 'payprice'].sum()
    #advertiser['cost_click'] = df.groupby(groupby_list).apply(click_cost)
    advertiser['eCPC'] = (advertiser['cost']/advertiser['num_clicks'])/1000
    advertiser['CPM'] = (advertiser['cost']/advertiser['num_impressions'])
    advertiser['CTR'] = (advertiser['num_clicks']/advertiser['num_impressions'])*100
    return advertiser.reset_index()

def plot_categorical(x, y, advertisers, data, join=True):
    fig, ax = plt.subplots(1, 1, figsize=(12, 12))
    sns.pointplot(x=x, y=y, hue="advertiser", data=data[data['advertiser'].isin(advertisers)], ax=ax, join=join)
    return fig, ax

def plot_factor(x, y, advertisers, data):
    fig, ax = plt.subplots(1, 1, figsize=(12, 12))
    sns.factorplot(x=x, y=y, hue="advertiser", data=data[data['advertiser'].isin(advertisers)], ax=ax, kind='bar')
    return fig, ax

def plot_bar(x, y, advertisers, data):
    fig, ax = plt.subplots(1, 1, figsize=(12, 12))
    sns.barplot(x=x, y=y, hue="advertiser", data=data[data['advertiser'].isin(advertisers)], ax=ax)
    return fig, ax
```
